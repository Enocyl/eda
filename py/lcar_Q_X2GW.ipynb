```python
def sch2verilog(path):
    sch_fns = [f for f in os.listdir(path) if os.path.isfile(os.path.join(path,f)) and f.split('.')[1] == 'sch'] 
    vfns = [fn.split('.')[0] for fn in sch_fns]
    ttl = len(sch_fns)
    pat = 'No Error found'
    mat = re.compile(pat)
    fail = 0
    for i,fn in enumerate(vfns):
        cmd_seg = ['sch2hdl -intstyle ise -family aspartan6 -verilog ']
        vfn = [os.path.join(path,fn)]
        cmd = ''.join(cmd_seg + vfn + ['.vf'] + [' -w '] + vfn + ['.sch'])
        rtn = None
        with os.popen(cmd,'r') as f:
            rtn = f.read() 
        if not mat.search(rtn):
            fail += 1
        print('\rSCH2HDL: {}\{}, Fail: {}'.format(i+1,ttl,fail),end='')
    print()
    print('**Process Finished**')
```

```python
def pmv_only_comment(fpath):
    ptn = r'(^[A-Z]\w+)\s+[#|\w+]'
    mch = re.compile(ptn)
    #pt2 = r'o?f\d*?d\d'
    #mc2 = re.compile(pt2)
    ptn2 = r'^\(\*'
    mch2 = re.compile(ptn2)
    pmvs = []
    data = None
    fname = os.path.split(fpath)[-1].split('.')[0]
    #print(fname,':')
    #with open(fpath,'r',encoding='utf-8') as f:
    with open(fpath,'r') as f:
        data = f.readlines()
    #print(data)
    #pmvs = mch.findall(''.join(data))
    pmv_endline = 0
    for idx,line in enumerate(data):
        if pmv_endline > idx:
            continue
        elif mch.match(line.strip()):
            pmvs.append(line.strip().split()[0])
            pmv_endline = get_pmv_endline(data,idx)
            #print('idx: {} - line: {} - endline: {}'.format(idx,data[idx],pmv_endline))
            for i in range(idx,pmv_endline+1):
                #print('idx: {} - v: {}'.format(i,data[i]))
                data[i] = '//' + data[i]
        elif mch2.match(line.strip()):
            #print('Macth: {} - {}'.format(idx,data[idx]))
            data[idx] = '//' + data[idx]
    return data, pmvs
    #with open(fpath,'w',encoding='utf-8') as f:
    #    f.writelines(data)
```

```python
def module_rename(vdata):
    ptn = r'_MUSER\w+'
    mch = re.compile(ptn)
    new_data = mch.sub('',vdata)   
    return new_data
```

```python
def comment_dnn(vdata):
    repl = r'//\s+`default_nettype none'
    mch1 = re.compile(repl)  
    ptn2 = r'`default_nettype none'
    mch2 = re.compile(ptn2)
    
    if mch1.search(vdata):
        return vdata
    elif mch2.search(vdata):
        #return mch2.sub(repl,vdata)
        return vdata.replace(ptn2,'//`default_nettype none')
    else:
        return vdata
```

```python
def vf2v(fpath,wfpath):
    fname = os.path.split(fpath)[-1].split('.')[0]
    ptn = 'module\s+' + fname
    mch = re.compile(ptn)
    data = None
    p0 = 0
    #for i,line in enumerate(open(fpath)):
    #    if mch.search(line):
    with open(fpath,'r',encoding='utf-8') as f:
        data = f.read()
    p0 = mch.search(data).span()[0]
    wdata = data[p0:]
    opath = os.path.join(wfpath,fname+'.v')
    with open(opath,'w',encoding='utf-8') as f:
        f.write(wdata)
    print('**P0: {} - File: {} Done**'.format(p0,fname+'.v'))
```
