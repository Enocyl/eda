```python
import scapy.all as scapy

def scan(ip):
    packet1 = scapy.ARP(pdst=ip)
    etherpacket = scapy.Ether(dst = 'ff:ff:ff:ff:ff:ff')

    broadcast_packet = etherpacket/packet1
    ans, unans = scapy.srp(broadcast_packet, timeout=10)

    print(ans.summary())

scan("192.168.1.1/24")
```

```python
def getHeader(fpath):
    data = []
    header = []
    with open(fpath,'rb') as f:
        data = f.readlines()[0][:64]
    #data = data.swapcase()
    for i in range(len(data)//4):
        header.append(bytes.hex(data[i*4:(i+1)*4]))
    return header
```

```python
header_path = './bitgen/std/header.bin'
```

```python
def headerGen(header_path, opcode_r='03', mboot_addr='400000', golden_addr='010000'):
    data = []
    with open(header_path,'rb') as f:
        data = f.readlines()
    #ld = []
    #for d in data:
    #    ld.append(len(d))
    #print('Before Modifed: $Doc_len - {}'.format(sum(ld)))  
    
    header = data[0][:64]
    header = b''.join([header[:26],
                       bytes.fromhex(mboot_addr[2:]),
                       header[28:30],
                       bytes.fromhex(opcode_r),
                       bytes.fromhex(mboot_addr[:2]),
                       header[32:34],
                       bytes.fromhex(golden_addr[2:]),
                       header[36:38],
                       bytes.fromhex(opcode_r),
                       bytes.fromhex(golden_addr[:2]),
                       header[40:64]])
        
    data[0] = b''.join([header,data[0][64:]])
    
    with open(header_path, 'wb') as f:
        f.writelines(data)
        
    with open(header_path,'rb') as f:
        data = f.readlines()
    ld = []
    for d in data:
        ld.append(len(d))
    #print('After Modifed: $Doc_len - {}'.format(sum(ld)))   
    
    header = data[0][:64]
    ind = []
    for i in range(len(header)//4):
        ind.append(bytes.hex(header[i*4:(i+1)*4]))        
    print(ind)    
```
